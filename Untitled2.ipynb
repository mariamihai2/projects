```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
fig, axes = plt.subplots(1, 3, figsize=(12, 3))

params = [(1, 100), (0.1, 10000), (0.01, 1000000)]
for ax, (step, n) in zip(axes, params):
    steps = np.random.choice([-1, 1], size=n)
    walk = np.cumsum(steps) * step
    ax.plot(np.linspace(1, n, n), walk, linewidth=0.5)
    ax.set_title(f"step={step}, n={n}")
    ax.set_ylim(-30, 30)
plt.savefig("scaling.png", dpi=150)
```
